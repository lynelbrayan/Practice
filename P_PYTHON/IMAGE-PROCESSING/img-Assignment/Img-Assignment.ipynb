
# shrink

# rotate 60%

```python

img = Image.open('Weird-Cat-Behaviors-Explained-6-1.jpeg')

img=img.rotate(60)

img.show()
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

# Translation

```python
img = cv2.imread('Weird-Cat-Behaviors-Explained-6-1.jpeg')

height, width = img.shape[:2]
  
quarter_height, quarter_width = height / 4, width / 4
  
T = np.float32([[1, 0, quarter_width], [0, 1, quarter_height]])

img_translation = cv2.warpAffine(img, T, (width, height))
  
cv2.namedWindow('original', cv2.WINDOW_KEEPRATIO)
cv2.namedWindow('translation', cv2.WINDOW_KEEPRATIO)  
cv2.imshow("original", img)
cv2.imshow('translation', img_translation)
cv2.waitKey(10000)
```

# edge detection

```python
loaded_image = cv2.imread('Weird-Cat-Behaviors-Explained-6-1.jpeg')
loaded_image = cv2.cvtColor(loaded_image,cv2.COLOR_BGR2RGB)
```

```python

plt.figure(figsize=(10,10))
plt.imshow(loaded_image,cmap="gray")
plt.axis("off")
plt.show()
```

```python

gray_image = cv2.cvtColor(loaded_image,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10,10))
plt.imshow(gray_image,cmap="gray")
plt.axis("off")
plt.show()
```

```python

edged_image = cv2.Canny(gray_image, threshold1=30, threshold2=100)

plt.figure(figsize=(10,10))
plt.imshow(edged_image,cmap="gray")
plt.axis("off")
plt.show()
```
