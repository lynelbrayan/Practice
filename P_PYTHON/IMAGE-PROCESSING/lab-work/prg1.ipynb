```python
# R. No: 2117014
# Program Number:10
# P Name: Demonstrate edge detection of your image with first order and second order derivative algorithms.
# Date: 07-09-2022 


#firstorder (canny,sobel,prewitt,roberts)
import cv2
import numpy as np
img=cv2.imread("C:\MCA_AIMIT\MCA_S2\Image_Process\images\py_img.png")
img_gaussian = cv2.GaussianBlur(img,(3,3),0)
#canny
img_canny = cv2.Canny(img,100,200)
#sobel
img_sobelx = cv2.Sobel(img_gaussian,cv2.CV_8U,1,0,ksize=5)
img_sobely = cv2.Sobel(img_gaussian,cv2.CV_8U,0,1,ksize=5)
img_sobel = img_sobelx + img_sobely
#prewitt
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
img_prewittx = cv2.filter2D(img_gaussian, -1, kernelx)
img_prewitty = cv2.filter2D(img_gaussian, -1, kernely)
#roberts
kernela=np.array([[-1,0,1],[-1,0,1]],dtype=float)
kernelb=np.array([[-1,0,1],[0,0,0]],dtype=float)
x=cv2.filter2D(img_gaussian,cv2.CV_16S,kernela)
y=cv2.filter2D(img_gaussian,cv2.CV_16S,kernelb)
absX=cv2.convertScaleAbs(x)
absY=cv2.convertScaleAbs(y)
Roberts=cv2.addWeighted(absX,0.5,absY,0.5,0)
cv2.imshow("Original Image", img)
cv2.imshow("Canny", img_canny)
cv2.imshow("Sobel X", img_sobelx)
cv2.imshow("Sobel Y", img_sobely)
cv2.imshow("Sobel", img_sobel)
cv2.imshow("Prewitt X", img_prewittx)
cv2.imshow("Prewitt Y", img_prewitty)
cv2.imshow("Prewitt", img_prewittx + img_prewitty)
cv2.imshow("Roberts",Roberts)
cv2.waitKey(0)
```

```python
# b) Watershed
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img=cv2.imread("C:\MCA_AIMIT\MCA_S2\Image_Process\images\py_img.png")
plt.title("Original image")
plt.imshow(img)
plt.show()
b,g,r = cv.split(img)
rgb_img = cv.merge([r,g,b])
plt.title("rgb image")
plt.imshow(rgb_img)
plt.show()
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
ret, thresh = cv.threshold(gray,0,255,cv.THRESH_BINARY_INV+cv.THRESH_OTSU)

# noise removal
kernel = np.ones((3,3),np.uint8)
opening = cv.morphologyEx(thresh,cv.MORPH_OPEN,kernel, iterations = 2)
plt.title("image without Noise")
plt.imshow(opening)
plt.show()

# sure background area
sure_bg = cv.dilate(opening,kernel,iterations=3)
plt.title("sure background area")
plt.imshow(sure_bg)
plt.show()

# Finding sure foreground area
dist_transform = cv.distanceTransform(opening,cv.DIST_L2,5)
ret, sure_fg = cv.threshold(dist_transform,0.7*dist_transform.max(),255,0)
plt.title("sure foreground area")
plt.imshow(sure_fg)
plt.show()

# Finding unknown region
sure_fg = np.uint8(sure_fg)
unknown = cv.subtract(sure_bg,sure_fg)
plt.title("unknown region")
plt.imshow(unknown )
plt.show()

# Marker labelling
ret, markers = cv.connectedComponents(sure_fg)
# Add one to all labels so that sure background is not 0, but 1
markers = markers+1
# Now, mark the region of unknown with zero
markers[unknown==255] = 0
markers = cv.watershed(img,markers)
img[markers == -1] = [255,0,0]
plt.title("Marker labelling ")
plt.imshow(markers )
plt.show()
```
