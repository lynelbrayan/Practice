```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
img1=cv2.imread("D:\Documents\images\m.png",cv2.IMREAD_GRAYSCALE)
img2=cv2.imread("D:\Documents\images\shanks.png",cv2.IMREAD_GRAYSCALE)
cv2.imshow("img1",img1)
cv2.imshow("img2",img2)
cv2.waitKey()
```

```python
directory = r'D:/images/'
try:
    os.mkdir(directory)
except Exception as e:
    print(e)

image = cv2.imread('D:\Documents\images\m.png')
cv2.imshow("Actual_Image",image)
intensity_matrix=np.ones(image.shape,dtype="uint8")*60
bi=cv2.add(image,intensity_matrix)
cv2.imshow("bright",bi)
di=cv2.subtract(image,intensity_matrix)
cv2.imshow("dark",di)
cv2.waitKey()

cv2.imwrite(directory+"bright.jpg",bi)
cv2.imwrite(directory+"dark.jpg",di)
```

```python
image = cv2.imread('../test-images/rgb.jpg')
cv2.imshow('Original Image', image)
R,G,B=cv2.split(image)
cv2.imshow("Red",R)
cv2.imshow("Green",G)
cv2.imshow("Blue",B)
cv2.waitKey()
```

```python

image = cv2.imread('D:\Documents\images\m.png')
cv2.imshow('Original Image', image)
height,width=image.shape[:2]
resized_down = cv2.resize(image, (200,200), interpolation= cv2.INTER_LINEAR)
resized_up = cv2.resize(image, (500,500), interpolation= cv2.INTER_LINEAR)
cv2.imshow('Resized Down by defining height and width', resized_down)
cv2.imshow('Resized Up image by defining height and width', resized_up)
cv2.waitKey()
```

```python
image = cv2.imread('D:\Documents\images\m.png')
height, width = image.shape[:2]
quarter_height, quarter_width = height / 4, width / 4
T = np.float32([[1, 0, quarter_width], [0, 1, quarter_height]])
img_translation = cv2.warpAffine(image, T, (width, height))
cv2.imshow("Originalimage", image)
cv2.imshow('Translation', img_translation)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img1=cv2.imread("D:\Documents\images\shanks.png")
height,width=img1.shape[:2]
mat=cv2.getRotationMatrix2D((width/2,height/2),60,1)
img=cv2.warpAffine(img1,mat,(width,height))
plt.imshow(img)
```
